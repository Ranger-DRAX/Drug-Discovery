## **Importing libraries**

```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## **Search for Target protein**

### **Target search for "coronavirus"**

```python
# Target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
targets.loc[targets['target_type'] == 'SINGLE PROTEIN']# target type == SINGLE PROTEIN

```

### **Select and retrieve bioactivity data for *SARS coronavirus 3C-like proteinase* (fifth entry)**

We will assign the fifth entry (which corresponds to the target protein, *coronavirus 3C-like proteinase*) to the ***selected_target*** variable 

```python
tg= targets[targets['target_chembl_id']=="CHEMBL3927"]
tg
```

```python
selected_target =  targets.target_chembl_id[6]
print(selected_target)
# (CHEMBL3927): This is the target protein, a key enzyme in coronaviruses.
# In drug discovery, this protein is studied to find compounds that can inhibit its function,
# potentially stopping the virus from replicating.
```

Here, we will retrieve only bioactivity data for *coronavirus 3C-like proteinase* (CHEMBL3927) that are reported as **IC50** values in nM (nanomolar) unit.

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

## **Data pre-processing of the bioactivity data**

### **Labeling compounds as either being active, inactive or intermediate**
The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be **active** while those greater than 10,000 nM will be considered to be **inactive**. As for those values in between 1,000 and 10,000 nM will be referred to as **intermediate**. 

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")

print(bioactivity_class.count("active"))
print(bioactivity_class.count("inactive"))
print(bioactivity_class.count("intermediate"))
```

### **Iterate the *molecule_chembl_id* to a list**

```python
df2.molecule_chembl_id
```

```python
mol_cid = []
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
  
mol_cid
```

### **Iterate *canonical_smiles* to a list**

```python

# canonical_smiles:  unique representation of a molecule's structure
#For ethanol (Câ‚‚Hâ‚…OH), the canonical SMILES string might be: CCO
canonical_smiles = []
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
df2[["canonical_smiles"]].head(10)
```

### **Iterate *standard_value* to a list**

```python
standard_value = []
for i in df2.standard_value:
  standard_value.append(i)
```

### **Combine the 4 lists into a dataframe**

```python

df3 = df2[['molecule_chembl_id', 'canonical_smiles', 'standard_value']]
df3
```

```python
df4 = pd.concat([df3, pd.Series(bioactivity_class, name='Bioactivity_Class')], axis=1)

#column wise concatenation
df4
```

Saves dataframe to CSV file

```python
df4.to_csv('bioactivity_data_preprocessed.csv', index=False)
df4.head(10)
```
